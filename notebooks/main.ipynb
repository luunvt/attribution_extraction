```python
dict_right = defaultdict(defaultdict)

for row in dataframe.iterrows():
    id, series = row
    asin = series['asin']
    title = series['product_title']
    des = series['description']
    attribute = dict_atribute[asin]
    
    dict_right[asin]['title'] = title
    dict_right[asin]['description'] = des
    dict_right[asin]['label'] = []

    for kv in attribute:
        for evidence in kv['evidences']:
            value = evidence['value']
            bg = evidence['begin']
            end = evidence['end']
            if value == title[bg:end]:
                dict_right[asin]['label'].append(evidence)
            elif value == title[bg-1:end-1]:
                dict_right[asin]['label'].append(evidence)
```

```python
# https://subscription.packtpub.com/book/data/9781838987312/2/ch02lvl1sec16/extracting-subjects-and-objects-of-the-sentence
import spacy

def get_subject_phrase(doc):
    for token in doc:
        if ("subj" in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]
        
def get_key_value_v1(bard_answer):
    nlp = spacy.load("en_core_web_sm")

    list_evidences = tuple(['(mentioned in "Product Title")', '(mentioned in title)', '(mentioned in title and "About this item")', '(mentioned in "Product Title" and "About this item")', '(mentioned in "About this item")'])

    lines = bard_answer.split("\n")
    correct_lines = [x for x in lines if x.endswith(list_evidences)]
    correct_lines_colon = []
    correct_lines_no_colon = []

    for x in correct_lines:
        if ":" in x:
            correct_lines_colon.append(x)
        else:
            correct_lines_no_colon.append(x)
    print(correct_lines_no_colon)
    result_dict =  {}
    for line in correct_lines_colon:
        line = line.replace("*", "").strip() 
        key = line.split(":")[0].strip()
        value = "".join(line.split(":")[1:]).replace('(mentioned in "About this item")', "").replace('(mentioned in title)', "").replace('(mentioned in title and "About this item")', "").strip()
        result_dict[key] = value
    for line in correct_lines_no_colon:
        line = line.replace("*", "").replace('(mentioned in "About this item")', "").replace('(mentioned in title)', "").replace('(mentioned in title and "About this item")', "").strip()
        doc = nlp(line)
        subject = get_subject_phrase(doc)
        if subject:
            result_dict[str(subject)] = "Yes"
    return result_dict
```

```python
from transformers import (
    TokenClassificationPipeline,
    AutoModelForTokenClassification,
    AutoTokenizer,
)
from transformers.pipelines import AggregationStrategy
import numpy as np

# Define keyphrase extraction pipeline
class KeyphraseExtractionPipeline(TokenClassificationPipeline):
    def __init__(self, model, *args, **kwargs):
        super().__init__(
            model=AutoModelForTokenClassification.from_pretrained(model),
            tokenizer=AutoTokenizer.from_pretrained(model),
            *args,
            **kwargs
        )

    def postprocess(self, all_outputs):
        results = super().postprocess(
            all_outputs=all_outputs,
            aggregation_strategy=AggregationStrategy.SIMPLE,
        )
        return np.unique([result.get("word").strip() for result in results])
```
