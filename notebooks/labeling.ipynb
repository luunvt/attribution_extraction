```python
import ast
import re
from collections import Counter

def get_label(descriptions, dataframe):
    cnt = Counter()
    asins = []
    sentences = []
    locs = []
    word = [] 

    for i, description in enumerate(descriptions):
        asin = dataframe.iloc[i]['asin']
        if str(description) == 'nan':
            continue

        description = ast.literal_eval(description)
        for sentence in description:
            locs_temp = []
            words_temp = []
            for material in materials_list:
                material = " " + material
                bgs = [m.start() for m in re.finditer(material.lower(), sentence.lower())]
                for bg in bgs:
                    pos_bg = bg + 1
                    pos_end = bg + len(material)
                    locs_temp.append([pos_bg, pos_end, 'MAT'])
                    words_temp.append(sentence[pos_bg: pos_end])
                    cnt[sentence[pos_bg: pos_end].lower()] += 1
            if len(locs_temp) > 0:
                asins.append(asin)
                sentences.append(sentence)
                locs.append(locs_temp)
                word.append(words_temp)
    print(len(asins), len(sentences), len(locs), len(word))
    return asins, sentences, locs, word, cnt
```

```python
list_file = ['../data/final_luu_a/final_luu/list_asin_1941.xlsx',
             '../data/final_luu_a/final_luu/list_asin_1981.xlsx',
             '../data/final_luu_a/final_luu/list_asin_2143.xlsx']

list_data = pd.concat([pd.read_excel(file) for file in list_file])
descriptions = list_data['description'].to_list()

asins, sentences, locs, word, cnt = get_label(descriptions, list_data)

output_data = pd.DataFrame()
output_data['asin'] = pd.Series(asins)
output_data['sentence'] = pd.Series(sentences)
output_data['locs'] = pd.Series(locs)
output_data['words'] = pd.Series(word)
```
