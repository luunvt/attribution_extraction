```python
import os
import pandas as pd
import ast
from collections import defaultdict
```

```python
folder_path = "/home/tanluuuuuuu/Desktop/luunvt/direct_indirect/data/final_luu_a/final_luu"

list_file = os.listdir(folder_path)
list_file
```

```python
list_file = ['list_asin_1921.xlsx', 'list_asin_3226.xlsx', 'list_asin_3389.xlsx', 'list_asin_3563.xlsx']

set_attribute = []
dict_attribute = defaultdict(set)
for f in list_file:
    f_path = os.path.join(folder_path, f)
    dataframe = pd.read_excel(f_path)
    list_product_detail = dataframe['product_detail'].to_list()
    for product_detail in list_product_detail:
        if str(product_detail) == 'nan':
            continue
        detail = ast.literal_eval(product_detail)
        attributes = []
        for att in list(detail.keys()): 
            lower_att = att.lower()
            attributes.append(lower_att)
            dict_attribute[lower_att].add(detail[att])
        set_attribute.extend(attributes)
set_attribute = sorted(set(set_attribute))
```
